```python
df2 = pd.read_csv('/content/drive/MyDrive/인공지능/14week/test_SectionDataset.csv', names=['title', 'section'], header=0)
df2 = df2[ df2['section'] == 0 ]
print( df2.shape )
print( df2['title'][0])

new_string = df2['title'][0]
print( new_string )
oneItem = {'title': new_string, 'section': 0 }
oneItem = pd.DataFrame(oneItem, index=[0])
df3 = pd.concat([df1, oneItem], ignore_index=True)
print(df3.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

print( df3[['title']].head(3) )
count_vector = CountVectorizer( ngram_range=(1,3) )
c_vector_titles = count_vector.fit_transform( df3['title'] )
print( c_vector_titles.shape )
print( c_vector_titles[0] )
print( "add1: ")
print( c_vector_titles[0, 0:40] )
print( "add2: ")
print( c_vector_titles[0, 0], c_vector_titles[0, 1],  c_vector_titles[0, 33], c_vector_titles[0, 255])
print( c_vector_titles[1, 0], c_vector_titles[1, 1],  c_vector_titles[1, 2])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

titles_c_sim = cosine_similarity( c_vector_titles )
print( titles_c_sim )

titles_c_sim = titles_c_sim.argsort()[:, ::-1]
print( titles_c_sim )
```

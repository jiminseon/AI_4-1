```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from keras.callbacks import ModelCheckpoint

import pandas as pd
import numpy
import os
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('/content/drive/MyDrive/colab/deeplearning/dataset/sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()

e.fit(Y_obj)
Y = e.transform(Y_obj).astype(float)
print(Y)
# 5개의 파일로 쪼갬
n_fold = 5
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

# 빈 accuracy 배열
accuracy = []

# 모델 저장 폴더 설정
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
   os.mkdir(MODEL_DIR)

# 모델의 설정, 컴파일, 실행
for train, test in skf.split(X, Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])

    model.fit(X[train], Y[train], epochs=100, batch_size=5)

    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)
    model.save("./model/lab7-model2-%.f.hdf5"%(accuracy.index(k_accuracy)))
# 결과 출력
print("\n %.f fold accuracy:" % n_fold, accuracy)
max =accuracy[0]
for i in range(n_fold):
  if (accuracy[i] >= max ):
    max = accuracy[i]
    index = i
print("Max index: ", index)
```

```python

from keras.models import load_model

from sklearn.model_selection import train_test_split
import os


X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=seed)

X_test = X[:5, :]
Y_test = Y[:5]

del model
model = load_model("./model/lab7-model2-%.f.hdf5"%(index))
model.evaluate(X_train, Y_train)

#예측 함수
predictions = model.predict(X_test)

# 실제 값, 예측 값 출력
for i in range(5):
  print("index: %d, actual: %.f, expect: %0.2f" %(i, Y_test[i], predictions[i]))
```

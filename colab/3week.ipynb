## ì‹¤ìŠµê³¼ì œ 3

```python
age_mean = round(df.age.mean(),2)
print('Average age:', age_mean)
low_age = df[df['age'] < age_mean]
high_age = df[df['age'] >= age_mean]
```

```python
low_age.thickness.plot.hist(bins=50, alpha=0.2)
high_age.thickness.plot.hist(bins=50, alpha=0.5)
plt.legend(['low_age', 'high_age'], loc='upper right')
plt.xlabel('thickness')
plt.show()
```

```python
import requests

url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=102"
#url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=101#&date= 00:00:00&page=2"

#크롤링 자동 방지 해제
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
           AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}
result = requests.get(url, headers=headers)
#print(result)

from bs4 import BeautifulSoup
soup = BeautifulSoup(result.text, 'html.parser')
#print(soup)

#news_titles = soup.find_all(class_='sa_test_title')
news_titles = soup.select('.sa_text_title')
#print(news_titles)

print('총 뉴스 제목 개수: ', len(news_titles))
for item in news_titles:
    print(item.get_text())
    #print(item['href'])
```

```python
import pandas as pd
import os

if os.path.isfile('./test_SectionDataset.csv'):
    df = pd.read_csv('./test_SectionDataset.csv', names=['title', 'section'], header=0)
else:
    bodyStruc = {'title': [], 'section': []}
    df = pd.DataFrame( bodyStruc, columns=['title', 'section'] )

for item in news_titles:
    texts = item.get_text()
    texts = texts.replace(',', "")
    texts = texts.replace('\'', "")
    texts = texts.replace('\"', "")
    texts = texts.replace('…', " ")
    print(texts)
    oneData = {'title': texts, 'section': '0'}
    new_oneData = pd.DataFrame(oneData,index=[0])
    df = pd.concat( [df, new_oneData], ignore_index=True )

df.to_csv('./test_SectionDataset.csv', encoding='UTF-8', index=False, header=True)
```
